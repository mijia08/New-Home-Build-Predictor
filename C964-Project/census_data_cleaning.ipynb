```python
import pandas as pd
```

```python
df = pd.read_csv('Datasets/Residential_Construction_Permits_by_County.csv')
df
```

```python
df = df[df['STUSAB'] == 'TN']
df = df.drop_duplicates(subset=['NAME'])
df = df.fillna(0)
df
```

```python
year_cols = [col for col in df.columns if col.startswith('ALL_PERMITS_')]
keep_cols = ['OBJECTID', 'GEOID', 'STATE', 'COUNTY','NAME', 'STUSAB', 'STATE_NAME']
to_keep = keep_cols + year_cols
df = df[to_keep]
df
```

```python
df.to_csv('Datasets/Cleaned_TN_Building_Permits_Census.csv', index=False)
```
